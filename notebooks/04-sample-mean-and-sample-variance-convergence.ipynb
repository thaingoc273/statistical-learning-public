```python
# code to illustrate the convergence of sample mean and sample variance as the sample size increases
# the underlying distribution is Bernoulli with p=0.5

import numpy as np
import matplotlib.pyplot as plt

# Define the population mean and standard deviation
pop_mean = 0
pop_std = 1

# Generate a list of sample sizes
max_sample_size = 500

sample = np.random.normal(pop_mean, pop_std, max_sample_size)
# Generate a list of sample means and sample variances for each sample size
sample_means = []
sample_variances = []

for n in range(1, max_sample_size):

    # Calculate the sample mean and sample variance
    sample_mean = np.mean(sample[:n])
    sample_variance = np.var(sample[:n])

    # Add the sample mean and sample variance to the respective lists
    sample_means.append(sample_mean)
    sample_variances.append(sample_variance)

# Plot the convergence of the sample mean and sample variance
plt.plot(sample_means, '-', label='Sample Mean')
plt.plot(sample_variances, '-', label='Sample Variance')
plt.xlabel('Sample Size')
plt.ylabel('Convergence')
plt.legend()
plt.show()
```
